```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

#np.random.seed(1)
x = np.random.randn(100000)
#print(x)

plt.figure()
#plt.plot(x)
plt.hist(x)
plt.show()
```

```python
np.random.seed(0)
x = np.random.randn(5)
y = np.random.randn(5)
plt.figure()
plt.plot(x,color="yellow", linewidth=10, linestyle=":")
plt.plot(y)
plt.xlabel("My X label", color='blue',size=18)
plt.ylabel("My Y label", color='red', size =18)
plt.title("My Title", color='purple',size =24)
#plt.legend("Data A", "Data B")
plt.xticks([0,1,2,3,4])
ax = plt.gca()
ax.set_xticklabels(["aa","bb","AaA",'aA',"XYZ"])
plt.show()
```

```python
np.random.seed(0)

x = np.random.randn(50)
y = np.random.randn(50)

plt.figure()
plt.scatter(x,y, color = (1,0,0), s = 500, alpha = 0.3)
plt.show()
```
