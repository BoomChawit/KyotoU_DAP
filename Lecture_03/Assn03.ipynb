### Figure 1

Use `data-fig1.npy` for this figure.

### Figure 2

Use `data-fig2.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

x = np.load("./data-Assn03/data-fig2.npy")
a = x[:,0]
b = x[:,1]
c = x[:,2]
plt.figure()
m = plt.plot(a, color = 'red')
n = plt.plot(b, color = 'green')
o = plt.plot(c, color = 'blue')

plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend([m[0],n[0],o[0]], ('red','green','blue'))        
plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a = np.arange(5)
b = np.load("./data-Assn03/data-fig4.npy")
plt.figure()
plt.bar(a,b , color = 'silver')

plt.xticks([0,1,2,3,4])
ax = plt.gca()
ax.set_xticklabels(["A","B","C",'D',"E"])
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use the "RdBu" colormap.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a = np.load("./data-Assn03/data-fig5.npy")
plt.figure()
x = plt.pcolor(a, cmap = 'RdBu', vmin = -1.0, vmax = 1)
plt.colorbar(x)
plt.show()
```

### BONUS

Use `data-fig-bonus-X.npy`, `data-fig-bonus-Y.npy` and `data-fig-bonus-Z.npy` for this figure.

*Hint*: This is one of Matplotlib's online examples.

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
x = np.load("./data-Assn03/data-fig-bonus-X.npy")
y = np.load("./data-Assn03/data-fig-bonus-Y.npy")
z = np.load("./data-Assn03/data-fig-bonus-Z.npy")

cset = ax.contour(x, y, z, extend3d=True, cmap=cm.coolwarm)

ax.clabel(cset, inline=1)

plt.show()
#credit: https://matplotlib.org/3.1.1/gallery/mplot3d/contour3d_2.html#sphx-glr-gallery-mplot3d-contour3d-2-py
```
