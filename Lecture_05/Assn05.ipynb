```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

year = np.array([i for i in range(1,10)])
sale = np.array([6151,762,856,1063,1190,1298,1421,1440,1518])
adv = np.array([23,26,30,34,43,48,52,57,58])
#print(year)
#print(sale)
#print(adv)
fig,ax = plt.subplots(1,2, figsize=(15,6))
ax[0].scatter(year,adv,color = 'black')  
ax[0].set_title('Years and Advertising')
ax[0].set_xlabel('Years')
ax[0].set_ylabel('Advertising (million euros)')

ax[1].scatter(adv,sale,color = 'black')
ax[1].set_title('Advertising and Sales')
ax[1].set_xlabel('Advertising (million euros)')
ax[1].set_ylabel('Sales (million euros)')

print("Year and advertising:",stats.linregress(year, adv))
print("Advertising and sales:",stats.linregress(adv, sale))
print('Correlation for year and advertising is',np.corrcoef(year, adv)[0,1])
print('Correlation for year and sales is',np.corrcoef(adv,sale)[0,1])
plt.show()
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

cig = np.array([5,23,25,48,17,8,4,26,11,19,14,35,29,4,23])
lon = np.array([80,78,60,53,85,84,73,79,81,75,68,72,58,92,65])

#print(cig,lon,len(cig),len(lon))
plt.figure()
plt.scatter(cig,lon, color = 'black', label='Data')
plt.xlabel("Cigarettes (number per day)")
plt.ylabel("Longevity (years)")

print("Cigarettes and Longevity:",stats.linregress(cig, lon))
r = np.corrcoef(cig,lon)[0,1]
print("Correlation coefficeint=",r)
a, b        = np.polyfit(cig, lon, deg=1)
x0          = min(cig)  
x1          = max(cig)  
y0          = a*x0 + b   
y1          = a*x1 + b   
plt.plot([x0,x1], [y0,y1], color='r', label='Regression line')
plt.text(35, 55, 'r = %f' %r, color='black')
plt.legend()
plt.show()
```

## Dataset 3:  Outliers

Load the attached data file: **dataset3.csv**. This file contains two columns: $x$ and $y$. The data contain two outliers. The outliers can be identified qualitatively (i.e., by plotting the data and visually checking for outliers).

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

x = np.loadtxt("dataset3.csv", delimiter = ',')
a = x[:,0]
b = x[:,1]
#print(x)
plt.figure()
plt.scatter(a,b, color='red',label='Data with outliers')

g = np.array( [True]*b.size )
#print(g)
g[-3] = False  # boolean array (last value is False)
g[3] = False
#print(g)
anew = a[g]
bnew = b[g]

r = np.corrcoef(a,b)[0,1]
rnew = np.corrcoef(anew, bnew)[0,1]
print("r old=",r)
print("r new=",rnew)
plt.scatter(anew,bnew, color='blue', label='Data without outlier')

m, c        = np.polyfit(a, b, deg=1)
x0          = min(a)  
x1          = max(a)  
y0          = m*x0 + c   
y1          = m*x1 + c   
plt.plot([x0,x1], [y0,y1], color='r', label='Regline with outliers')
plt.text(-20, 4, 'r = %f' %r, color='black')

k, l        = np.polyfit(anew, bnew, deg=1)
x2          = min(anew)  
x3          = max(anew)  
y2          = k*x2 + l   
y3          = k*x3 + l   
plt.plot([x2,x3], [y2,y3], color='b', label='Reg line without outliers')
plt.text(-2, 11, 'r = %f' %rnew, color='black')

plt.legend()
plt.show()
```
